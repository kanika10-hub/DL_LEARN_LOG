tanh activation

```python
def tanh(x):
  return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(3)
```

```python
tanh(0)
```

```python
tanh(-1)
```

```python
tanh(1)
```

```python
 tanh(-100)
```

ReLu function

```python
def ReLu(x):
  return max(0,x)
```

```python
ReLu(100)
```

```python
 def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(10)
```

```python
leaky_relu(-10)
```
