```python
import numpy as np
y_predicted=np.array([1,1,0,0,1])
y_true=np.array([0.30,0.7,1,0,0.5])
```

```python
def mae(y_true,y_predicted):
  total_error=0
  for yt,yp in zip(y_true,y_predicted):
    total_error+=abs(yt-yp)
    print("total error:", total_error);
  mae=total_error/len(y_true)
  print("mae:",mae)
  return mae
```

```python
mae(y_true,y_predicted)
```

```python
np.mean(np.abs(y_predicted-y_true))
```

since log 0 in not defined and log 1 is exact 0 , that will not work in training , so we use epsilon , that is close to 0

clamping values between [epsilon,1-epsilon]

```python
np.log([0.000000000000001])
```

```python
epsilon=1e-15
```

```python
y_predicted_new=[max(i,epsilon) for i in y_predicted]
y_predicted_new
```

```python
y_predicted_new=[min(i,1-epsilon)for i in y_predicted_new]
y_predicted_new
```

```python
y_predicted_new=np.array(y_predicted_new)
np.log(y_predicted_new)
```

```python
-np.mean(y_true*np.log(y_predicted_new)+np.log(1-y_predicted_new))
```
